```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def phase(X,t,k):
    
    #q is theta
    #p is theta'
    
    q=X[0]
    p=X[1]
    
    dqdt= p
    dpdt= -((k)**2)*(np.sin(q))
    
    dXdt= np.array([dqdt, dpdt])
    
    return dXdt
```

```python
#initial condition of X
X0= np.array([0,5])

#time points
t= np.linspace(0,15,1000)
```

```python
#specify parameters

k=10

#solve ODE
X=odeint(phase, X0, t, args=(k,))
```

```python
plt.figure(figsize=(12,5))
plt.plot(t, X[:,0], 'b-', label='q')
plt.plot(t, X[:,1], 'g-', label= 'p')
plt.xlabel('Time')
plt.ylabel('value')
plt.legend(loc='best')
plt.title('Simple pendulum')
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(X[:,0], X[:,1], 'b-', label='proper')
plt.xlabel('Q')
plt.ylabel('P')
plt.legend(loc='best')
plt.title('Simple pendulum')
plt.show()
```
